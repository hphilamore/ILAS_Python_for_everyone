## Searching Files
Last week, we learnt to read information from a file.

The following line reads the file using the mode specifier `r`.

```python
with open("sample_data/poem.txt", "r" ) as file:

    for line in file:
        print(line, end="")
```

Instead of printing every line, we can use the keyword `in` to find strings:

```python
with open("sample_data/poem.txt", "r" ) as file:

    for line in file:
        
        # Is "the" in the line
        # print("the" in line)

        # Is "t" in the line
        # print("t" in line)
        
        # Only print the line if it contains "the"
        if "the" in line:
            print(line, end="")

```

For additional search functionality, we can use Python’s regular expression pattern matching module, `re`.

```python
import re
```

The above example can be repeated using the function `re.search()`, which is similar to `in`:  

```python
with open("sample_data/poem.txt", "r" ) as file:

    for line in file:

            # Is "the" in the line
            # print(re.search(r"the", line))

            # Is "t" in the line - boolean value
            print(bool(re.search(r"t", line)))

            # Only print the line if it contains "the"
            if re.search(r"the", line):
                print(line, end="")
```

##### What does this code do?
>`re.search(r"the", line)`

`search()` takes two arguments: 
- the string you want to search for 
- the string you want to search within 

The match string is just like a normal string. The r in front of the string is a label to tell Python not to expand slashes. 

The argument is returned as a *Python regular expression*.

>`bool(re.search(r"t", line))`

To get a True or False value, the expression is *cast* as a boolean value. 

By using `re` we can do things like making our search case insensitive.

```python
line = "My cat is brown"

# search for "cat"
cat = re.search( r"Cat", line )

# search for "cat", "CAT", "CaT", "caT" etc.
# cat = re.search( r"Cat", line, re.IGNORECASE )

print(bool(cat))
```

In the last class, we learnt to use the method `split`:

```python
# Prints indivual words as list
with open("sample_data/poem.txt", "r" ) as file:
    print(file.read().split())

print()

# Sequentially prints each element of list
with open("sample_data/poem.txt", "r" ) as file:
    for word in file.read().split() :
        print(word, end=" ")
```

### Replacing Words
As well as searching, it can be useful to replace words.

Last week, we learnt to replace words in an existing file with another word. 

```python
line = "Round the ragged rock the ragged rascal ran"
print(line)

# replace "ra" with "RA"
line = line.replace("ra", "RA")
print(line)
```

We can limit the number of instances replaced:

```python
line = "Round the ragged rock the ragged rascal ran"
print(line)

# replace only a maximum of 2 matches of "ra" with "RA"
line = line.replace("ra", "RA", 2)
print(line)
```

Again the `re` module allows you to perform more complex search-related operations.

Sometimes you may want to perform the substitution at a specific place in the line. 

```python
line = "Round the ragged rock the ragged rascal ran"

# replace all instances of "ra" with "RA"
print(re.sub(r"ro", "RA", line))

# replace only the first two instances of "ra" with "RA"
print(re.sub(r"ra", "RA", line, 2))

# replace "ra" with "RA" ONLY IF at start of line (use ^)
print(re.sub(r"^ra", "RA", line)) 

# replace "ra" with "RA" ONLY IF at end of line (use $)
print(re.sub(r"$ra", "RA", line))
```

To perform a case-insensitive replacement:

```python
line = "Round the ragged rock the ragged rascal ran"
print(line)


line = re.sub(r"ro", "RO", line, flags=re.IGNORECASE)
print(line)
```
